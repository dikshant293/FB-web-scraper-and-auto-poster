```python
def member_count(group_info):
    try:
        
        driver.get('https://mbasic.facebook.com'+group_info)
        time.sleep(2)
        pageContent = driver.page_source
        soup = BeautifulSoup(pageContent, 'html.parser')
        start_ind = soup.text.find('Members')
        start_ind=start_ind+7

        s=''
        while(soup.text[start_ind].isdigit()):
            s=s+soup.text[start_ind]
            start_ind+=1
        
        total_members = s
        title = soup.title.text
        

        print('https://www.facebook.com'+group_info+'about')
        driver.get('https://www.facebook.com'+group_info+'about')
        time.sleep(2)
        pageContent = driver.page_source
        soup2 = BeautifulSoup(pageContent, 'html.parser')
        
        start = soup2.text.find('Activity') + 8
        new_posts_today = soup2.text[start:].split(' ')[0]
        new_posts_today.replace('\n','')
        if(new_posts_today in ['','No']):
            new_posts_today = '0'
        
        start = soup2.text.find('today') + 5
        new_posts_last_month = soup2.text[start:].split(' ')[0]
        new_posts_last_month.replace('\n','')
        if(new_posts_last_month in ['','No']):
            new_posts_last_month = '0'
        if(soup2.text.find("This content")!=-1):
            new_posts_today='NA'
            new_posts_last_month='NA'
            
        url = 'https://www.facebook.com'+group_info+'about'
        print("{} total members\t\t{} posts today\t\t{} posts this month\t\t{}".format(total_members,
                                                                                    new_posts_today,
                                                                                    new_posts_last_month,
                                                                                    title))
        return total_members, new_posts_today, new_posts_last_month, title, url
    except:
        print('some error')
```

```python
data = pd.read_csv('delhidb.csv')
data.head()
```

```python
raw_list = list(data['Facebook group'])
raw_list
```

```python
link_list = []
for x in raw_list:
    if(type(x)==type('test')):
        x=x.split('\n')
        for y in x:
            y=y.split(",")
            for z in y:
                if(z.find('groups')!=-1 and z.find(' ')==-1):
                    new_link = z.strip()
                    ind = new_link.find('.com')+4
                    end_ind = new_link.find('?')
                    if(end_ind==-1):
                        link_list.append(new_link[ind:])
                    else:
                        link_list.append(new_link[ind:end_ind])
link_list           
```

```python
name=[]
members=[]
posts_today=[]
posts_last_month=[]
urls=[]

for i in link_list:
    total_members, new_posts_today, new_posts_last_month, title, url = member_count(i)
    name.append(title)
    members.append(total_members)
    posts_today.append(new_posts_today)
    posts_last_month.append(new_posts_last_month)
    urls.append(url)
    
final_data = pd.DataFrame({'Title':name,
                           'URL':urls,
                           'Total Members':members,
                           'Posts Today':posts_today,
                           'Posts Last Month':posts_last_month})
```
