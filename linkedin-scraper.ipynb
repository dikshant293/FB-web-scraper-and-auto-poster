```python
# imports
from selenium import webdriver
from time import sleep
from parsel import Selector
import pandas as pd
```

```python
# specifies the path to the chromedriver.exe
driver = webdriver.Chrome()

# driver.get method() will navigate to a page given by the URL address
driver.get('https://www.linkedin.com')

# locate email form by_class_name
username = driver.find_element_by_id('session_key')

# send_keys() to simulate key strokes
username.send_keys('usrnamehere')

# sleep for 0.5 seconds
sleep(0.5)

# locate password form by_class_name
password = driver.find_element_by_id('session_password')

# send_keys() to simulate key strokes
password.send_keys('passwordhere')
sleep(0.5)

# locate submit button by_xpath
sign_in_button = driver.find_element_by_xpath('//*[@type="submit"]')

# .click() to mimic button click
sign_in_button.click()
```

```python
from selenium.webdriver.common.keys import Keys

# specifies the path to the chromedriver.exe
driver2 = webdriver.Chrome()

# driver.get method() will navigate to a page given by the URL address
driver2.get('https://www.google.co.in')
sleep(1)
# locate search form by_name
search_query = driver2.find_element_by_name('q')

# send_keys() to simulate the search text key strokes
search_query.send_keys('site:linkedin.com/in/ AND "Consultant" AND "India"')

# .send_keys() to simulate the return key 
search_query.send_keys(Keys.RETURN)

# linkedin_urls = driver.find_elements_by_class_name('iUh30')
# linkedin_urls = [url.text for url in linkedin_urls]
# sleep(0.5)
# results = driver2.find_elements_by_css_selector('div.g')
# link = results[0].find_element_by_tag_name("a")
# href = link.get_attribute("href")
hrefs = []

for i in range(4):
    results = driver2.find_elements_by_css_selector('div.g')
    links = [x.find_element_by_tag_name("a") for x in results]
    hrefs += [x.get_attribute("href") for x in links]
    driver2.find_element_by_id('pnnext').click()
    
hrefs
```

```python
urldb = pd.DataFrame({'URL':hrefs})
urldb.to_csv('linkedin-urls.csv',index=False)
urldb
```
